Below is the code that I eventually got to work for scraping the data.

Here we are going to loop through and make multiple api requests to make our data. We are going to remove all ',' and lowercalse all the words as well as add the flavors to the short descriptions to have more words and descriptions of our different coffees.

```python
# Url of the request from Google Chrome network tab
url = 'https://92nybnyamc-dsn.algolia.net/1/indexes/coffee/query?x-algolia-agent=Algolia%20for%20JavaScript%20(3.35.1)%3B%20Browser&x-algolia-application-id=92NYBNYAMC&x-algolia-api-key=197eb4b9981532b5a33639b6b70c8a59'

# Headers from network tab
headers = {
    "accept": "application/json",
    "accept-language": "en-US,en;q=0.9",
    "content-type": "application/x-www-form-urlencoded",
    "sec-ch-ua": "\" Not A;Brand\";v=\"99\", \"Chromium\";v=\"96\", \"Google Chrome\";v=\"96\"",
    "sec-ch-ua-mobile": "?0",
    "sec-ch-ua-platform": "\"Windows\"",
    "sec-fetch-dest": "empty",
    "sec-fetch-mode": "cors",
    "sec-fetch-site": "cross-site"
}

# Needed a session to have the correct referrer
s = requests.Session()
s.headers.update({'referer': "https://www.mistobox.com/"})

# Store data and labels in arrays with matching indices
data = []
labels = []

# There are 55 "pages" of coffee so we call 55 times
for i in range(1, 55):
    # Make a post request
    resp = s.post(url, data="{\"params\":\"query=&filters=searchable%3Atrue&hitsPerPage=12&page=" + str(i) + "&facets=%5B%22*%22%5D\"}", headers=headers)
    for hit in resp.json()['hits']:
        # Format the label and the description nicely
        desc = hit['short_description'].replace(',', '')
        flavors = ' '.join([x.lower() for x in hit['flavors']])
        total_data = desc + ' ' + flavors
        
        lab = '-'.join([x.lower() for x in hit['roasts']])
        
        # Only add data if both label and description are present 
        if total_data != '' and lab != '':
            data.append(total_data)
            labels.append(lab)
```

```python
# Function to generate frequency graph of different words for roast of coffee
def generate_graph_for_words(words):
    graph = [0, 0, 0, 0, 0]
    possible_labels = ['light', 'light-medium', 'medium', 'medium-dark', 'dark']

    # Look through descriptions and labels and increment count for label if word is in description
    for x, y in zip(data, labels):
        for word in words:
            if word in x:
                graph[possible_labels.index(y)] += 1
                break

    # label axis and plot graoh
    fig = plt.figure(figsize=(20,10))
    plt.bar(possible_labels,graph)
    plt.ylabel(f'Frequency of Word{"s" if len(words) > 1 else ""} {", ".join([x.capitalize() for x in words])}', fontsize=20)
    plt.xlabel('Roast of Coffee', fontsize=20)
    plt.title(f'Frequency of Word{"s" if len(words) > 1 else ""} {", ".join([x.capitalize() for x in words])} vs Roast of Coffee', fontsize=25)
    plt.show()
```

The function above essentially counts how many coffee descriptions contain a specific word, partitioned by the roast of the coffee. This is a very important function as it can help to show if my predictions about different words and their trends are true.

```python
count = [0, 0, 0, 0, 0]
possible_labels = ['light', 'light-medium', 'medium', 'medium-dark', 'dark']

# Count how many times the different coffees appear
for x, y in zip(data, labels):
    count[possible_labels.index(y)] += 1

# Plot the findings
fig = plt.figure(figsize=(20,10))
plt.bar(possible_labels,count)
plt.ylabel(f'Frequency', fontsize=20)
plt.xlabel('Roast of Coffee', fontsize=20)
plt.title(f'Frequency vs Roast of Coffee', fontsize=25)
plt.show()
```

Ahh this now makes a lot of sense! We can see that medium coffee is way more popular on this website than other roasts of coffee so it appears more in our graphs! Lets try making that function again, but this time looking at the relative frequency.

```python
# Plot different words relative frequency for differernt roasts of coffee
def generate_graph_for_words_relative(words):
    graph = [0, 0, 0, 0, 0]
    possible_labels = ['light', 'light-medium', 'medium', 'medium-dark', 'dark']

    # Count how many times word appears for different roast
    for x, y in zip(data, labels):
        for word in words:
            if word in x:
                graph[possible_labels.index(y)] += 1
                break
    
    # Divide by total count for roast
    for i in range(len(graph)):
        graph[i] /= count[i]
    
    # Plot
    fig = plt.figure(figsize=(20,10))
    plt.bar(possible_labels,graph)
    plt.ylabel(f'Frequency of Word{"s" if len(words) > 1 else ""} {", ".join([x.capitalize() for x in words])}', fontsize=20)
    plt.xlabel('Roast of Coffee', fontsize=20)
    plt.title(f'Relative Frequency of Word{"s" if len(words) > 1 else ""} {", ".join([x.capitalize() for x in words])} vs Roast of Coffee', fontsize=25)
    plt.show()
```

```python
generate_graph_for_words_relative(['chocolate', 'cocoa'])
```

This is not perfect, but is a lot more like what I expected to see (and it is a nice looking graph). We can see that the words chocolate and cocoa appear most frequently in medium-dark coffee but also appear frequently in dark and medium coffee and even sometimes in light-medium and light coffee.

This graph has really improved. We can see that the word fruit almost certainly corresponds to a light or light-medium coffee which hopefully means that tf-idf or Doc2vec will perform well later.

Now that we have done some fun EDA, lets get into the really fun stuff, tfidf vs Doc2Vec. After looking at the graphs above, I am pretty concerned about their performance. I really thought that there was going to be specific words that more obviously corresponded to different roasts of coffee, but that did not exactly seem to be the case.

We are going to start with tf-idf. This essentially counts how many times a word appears in a document as well as how many documents the word appears in. The general idea is that words that appear in many documents are less important than words that appear in only a few documents. When evaluating the model on the test set, we will be using cosine similarity to find the most similar document from the test set and using that label as our answer.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Split into train and test for initial test
train_x, train_y = data[:400], labels[:400]
test_x, test_y = data[400:], labels[400:]

# Fit on test data
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(train_x).toarray()

# Now check accuracy of test data
correct = 0
for x, y in zip(test_x, test_y):
    test_vec = vectorizer.transform([x]).toarray()
    temp = X.dot(test_vec.T)

    for mat in temp.T:
        temp2 = [(train_y[i], x) for i, x in enumerate(mat)]
        if sorted(temp2, key=lambda i: i[1], reverse=True)[0][0] == y:
            correct += 1

print(f'accuracy = {correct/len(test_y)}')
```

This is really not too bad a result for tf-idf. I was expecting a lot higher initially, but after seeing the graphs for different words, I was not expecting that high of an accuracy.

Now lets try Doc2vec. I have higher hopes for this model after seeing how Word2vec is able to group words based on meaning. My hope is that Doc2vec is able to group documents based on coffee roast.

```python
# Need a few more dependencies here
!pip install gensim
!pip install nltk
from gensim.test.utils import common_texts
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')
```

```python
# Split into initial train/test data
train_x, train_y = data[:400], labels[:400]
test_x, test_y = data[400:], labels[400:]

# Train the Doc2Vec model
tagged_data = [TaggedDocument(words=word_tokenize(_d.lower()), tags=[str(i)]) for i, _d in enumerate(train_x)]
model = Doc2Vec(tagged_data)


# Now check accuracy of test data
correct = 0
for x, y in zip(test_x, test_y):
    test = word_tokenize(x)
    ivec = model.infer_vector(test)
    g = model.dv.most_similar(positive=[ivec], topn=1)
    if train_y[int(g[0][0])] == y:
        correct += 1

print(f'accuracy = {correct/len(test_y)}')
```

This accuracy is a little disappointing as I was expecting it to surpass tf-idf. However, this could just be an unlucky train/test split for Doc2vec. To really see which model is better, I am going to split the data 10 times randomly using a 2/3 train 1/3 test split. Then I will evaluate the accuracy for each iteration and average over the accuracies to get a better picture of these models' performances.

```python
from sklearn.model_selection import train_test_split
```

```python
acc_tfidf = []
acc_doc2vec = []

for i in range(10):
    # 2/3 1/3 split
    X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.33)
    
    # train and test tfidf
    vectorizer = TfidfVectorizer()
    X = vectorizer.fit_transform(X_train).toarray()

    # Now check accuracy of test data
    correct = 0
    for x, y in zip(X_test, y_test):
        test_vec = vectorizer.transform([x]).toarray()
        temp = X.dot(test_vec.T)

        for mat in temp.T:
            temp2 = [(y_train[i], x) for i, x in enumerate(mat)]
            if sorted(temp2, key=lambda i: i[1], reverse=True)[0][0] == y:
                correct += 1

    acc_tfidf.append(correct/len(test_y))
    
    # train and test doc2vec
    tagged_data = [TaggedDocument(words=word_tokenize(_d.lower()), tags=[str(i)]) for i, _d in enumerate(X_train)]
    model = Doc2Vec(tagged_data)


    # Now check accuracy of test data
    correct = 0
    for x, y in zip(X_test, y_test):
        test = word_tokenize(x)
        ivec = model.infer_vector(test)
        g = model.dv.most_similar(positive=[ivec], topn=1)
        if y_train[int(g[0][0])] == y:
            correct += 1

    acc_doc2vec.append(correct/len(test_y))

print(f'Average tfidf acc = {sum(acc_tfidf)/10}')
print(f'Average Doc2Vec acc = {sum(acc_doc2vec)/10}')
```

These are really not that great of performances, unfortunately. I am especially disappointed with Doc2vec's performance, however, it could be indicative of bad data for this type of model. Seeing that words like chocolate, which we would want to heavily correspond to a specific roast type, actually was spread almost evenly across 3 different roast types was not a good sign. In fact, I think if we had a classifier that predicted medium every time, it would be better than these for this specific data. Obviously, that would not generalize that well so it is a bad idea, but it is just something interesting to note.
